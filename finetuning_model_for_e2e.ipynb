```python
#filename = "data/e2e-dataset/trainset.csv"
#filename = "F:/DellGL3/School/CMU/classes/Fall2022/10708/Project/gpt2/e2e_nlg/data/e2e_nlg_train.txt"
def tokenize_e2e(loaded_corpus,filename):
    #add new chars to corpus
    with open(filename, 'r') as f:
        tokens = 0
        for line in f:
            words = list(line) + ['<eos>']
            tokens += len(words)
            for word in words:
                loaded_corpus.dictionary.add_word(word)

    with open(filename, 'r') as f:
        ids = torch.LongTensor(tokens)
        token = 0
        for line in f:
            words = list(line) + ['<eos>']
            for word in words:
                ids[token] = loaded_corpus.dictionary.word2idx[word]
                token += 1

    return ids

def tokenize_e2e_enwik(loaded_corpus,filename):
    #add new chars to corpus
    with open(filename, 'r') as f:
        tokens = 0
        for line in f:
            words = list(line) + ['<eos>']
            tokens += len(words)
            for word in words:
                #print("word: ",word)
                word_as_int = str(ord(word)) if word!='<eos>' else word
                loaded_corpus.dictionary.add_word(word_as_int)

    with open(filename, 'r') as f:
        ids = torch.LongTensor(tokens)
        token = 0
        for line in f:
            words = list(line) + ['<eos>']
            for word in words:
                word_as_int = str(ord(word)) if word!='<eos>' else word
                ids[token] = loaded_corpus.dictionary.word2idx[word_as_int]
                token += 1

    return ids
```
